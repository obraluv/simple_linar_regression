```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

insurance_data = pd.read_csv('insurance.csv') # implemented the pandas function to read our csv file
insurance_data.head() # to display the first five elements in our file, did this to manually visualise our data
```

```python
missing_values_count = insurance_data.isnull().sum() # to check for missing data

# Look at the number of missing points in all columns
missing_values_count[0:7]
```

```python
insurance_data.tail() # displays the last five elements, use this to further manually visualise the data
```

```python
x = insurance_data.loc[:, 'age'].values  # select all rows and age column
y = insurance_data.loc[:, 'charges'].values  # select all rows and charges column

x = x.reshape(-1, 1) # creates a 2D array for the x-axis, as scikit-learn's LinearRegression model expects

# plotting scatter plot
plt.scatter(x, y, color = 'b')
plt.title('Age/Charges')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
plt.close()
```

```python
# fit our model using linear regression
age_model = LinearRegression() 
age_model.fit(x,y)

y_pred = age_model.predict(x) # making predictions

# plotting the best-fit line
plt.scatter(x, y, color = 'b')
plt.plot(x, y_pred, color = 'r')
plt.title('Age/Charges')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

From the scatter plots above we can see that there is a positive correlation between age and charges because the data points show a general upward trend from left to right. This means that as age increases, charges tend to increase as well
